```python
import pandas as pd
from sqlalchemy import create_engine
import requests
import time
import os
import datetime as dt
```

```python
#list of query parameters:

query_list=['launchpads', 'capsules']
```

```python
#list of all the output corresponding to the query parameters
capsules=[]
launchpads=[]
output=[capsules,launchpads]
```

```python
url = f'https://api.spacexdata.com/v3/'
#example url
```

```python
#output list of all the requests

output_list=[]

for item in query_list:
    request = requests.get(f'{url+ item}').json()
    
    output_list.append(request)
```
